```python
import cv2
import dlib

img = cv2.imread('/content/drive/MyDrive/kinship project/input/Ignoti/23Aprile[5].png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

detector = dlib.get_frontal_face_detector()

faces = detector(img_gray, 1)


print(faces)
print("Numero di volti rilevati: ", len(faces))


i = 0
for face in faces:
    x1 = face.left()
    y1 = face.top()
    x2 = face.right()
    y2 = face.bottom()

    # Disegna un rettangolo intorno al volto
    #cv2.rectangle(img, (x1,y1), (x2,y2),(0,255,0),3)
    crop_img = img[y1:y2, x1:x2]
    i = i + 1
    cv2.imwrite("/content/drive/MyDrive/kinship project/input/Ignoti/test/23Aprile[5]"+str(i)+".png", crop_img)
    
```
