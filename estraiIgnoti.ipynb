```python
import cv2
import dlib

img = cv2.imread('/content/drive/MyDrive/kinship_prediction/input/Ignoti/23Aprile[5].png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

detector = dlib.get_frontal_face_detector()

faces = detector(img_gray, 1)

#printing the coordinates of the bounding rectangles
print(faces)
print("Number of faces detected: ", len(faces))

# Using a for loop in order to extract the specific coordinates (x1,x2,y1,y2)
i = 0
for face in faces:
    x1 = face.left()
    y1 = face.top()
    x2 = face.right()
    y2 = face.bottom()

    # Drawing a rectangle around the face detected
    #cv2.rectangle(img, (x1,y1), (x2,y2),(0,255,0),3)
    crop_img = img[y1:y2, x1:x2]
    i = i + 1
    cv2.imwrite("/content/drive/MyDrive/kinship_prediction/input/Ignoti/test/23Aprile[5]"+str(i)+".png", crop_img)
    
```
